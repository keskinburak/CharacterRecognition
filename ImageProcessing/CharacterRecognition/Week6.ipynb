```python
# Bubble Sort
def bubble(list):
    length = len(list) - 1
    sorted = False

    while not sorted:
        sorted = True
        for i in range(length):
            if list[i] > list[i+1]:
                sorted = False
                list[i], list[i+1] = list[i+1], list[i]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Character Image Added to Array for loop
array=["a1.jpg","a2.jpg","a3.jpg","a4.jpg","a5.jpg","b1.jpg","b2.jpg","b3.jpg","b4.jpg","b5.jpg","c1.jpg","c2.jpg","c3.jpg","c4.jpg","c5.jpg","d1.jpg","d2.jpg","d3.jpg","d4.jpg","d5.jpg","e1.jpg","e2.jpg","e3.jpg","e4.jpg","e5.jpg",]

# Array2 for reshaped Image Array
array2=[]
for k in range(25):
    img1 = plt.imread(array[k])
    img1.ndim
    img1.shape
    img3 = np.zeros(img1.shape[0:2])
    img3.shape
    threshold = 120
    
    #Turning Gray Level
    for i in range(img1.shape[0]):
        for j in range(img1.shape[1]):
            n = img1[i,j,0]/3 + img1[i,j,1]/3 + img1[i,j,2]/3
            if n > threshold:
                img3[i,j] = 255
            else:
                img3[i,j] = 0
    plt.imshow(img3, plt.cm.gray)
    plt.show()
    
    #Reshape Image and Added to Array
    img3=np.array(img3,np.int32).reshape(1,10000)
    array2.append(img3)
```

```python
#Variable for Characters Centers

CenterA=np.zeros((1,10000))
CenterB=np.zeros((1,10000))
CenterC=np.zeros((1,10000))
CenterD=np.zeros((1,10000))
CenterE=np.zeros((1,10000))


for i in range(0,25):
    if i<5:
        CenterA=CenterA+array2[i]
    elif i<10:
        CenterB=CenterB+array2[i]
    elif i<15:
        CenterC=CenterC+array2[i]
    elif i<20:
        CenterD=CenterD+array2[i]
    else:
        CenterE=CenterE+array2[i]
        
#Final State of Centers
CenterA=CenterA/5
CenterB=CenterB/5
CenterC=CenterC/5
CenterD=CenterD/5
CenterE=CenterE/5

#Test Image Added to Process
test=plt.imread("test.jpg")
test.ndim
test.shape
threshold=120

test1= np.zeros(test.shape[0:2])
test1.shape

#Test Turning Gray Level
for i in range(test.shape[0]):
        for j in range(test.shape[1]):
            n = test[i,j,0]/3 + test[i,j,1]/3 + test[i,j,2]/3
            if n > threshold:
                test1[i,j] = 255
            else:
                test1[i,j] = 0
                
                
#Reshaped for the Distance Process
test1=np.array(test1,np.int32).reshape(1,10000)

#Calculated Distance for Recognition
distA = np.linalg.norm(CenterA-test1)
distB = np.linalg.norm(CenterB-test1)
distC = np.linalg.norm(CenterC-test1)
distD = np.linalg.norm(CenterD-test1)
distE = np.linalg.norm(CenterE-test1)

array3=[]
array3.append(distA)
array3.append(distB)
array3.append(distC)
array3.append(distD)
array3.append(distE)

bubble(array3)

#Says Recognized Character
if array3[0]==distA:
    print("Yazdığınız Karakter a")
elif array3[0]==distB:
    print("Yazdığınız Karakter b")
elif array3[0]==distC:
    print("Yazdığınız Karakter c")
elif array3[0]==distD:
    print("Yazdığınız Karakter d")
else :
    print("Yazdığınız Karakter e")



                
```
