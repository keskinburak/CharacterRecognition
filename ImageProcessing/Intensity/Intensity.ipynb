```python
import numpy as np
import matplotlib.pyplot as plt

def RGBtoGrayLevel(image):
    img2=np.zeros((image.shape[0:2]))
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            img2[i,j]=sum(image[i,j,:])/3
    return img2

def redIntensity(img):
    img2 = np.zeros((img.shape[0], img.shape[1], img.shape[2]), dtype=np.uint8)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if(img[i,j,0]+50 > 255):
                img2[i,j,0] = 255
            else:
                img2[i,j,0] = img[i,j,0]+50
            img2[i,j,1] = img[i,j,1]
            img2[i,j,2] = img[i,j,2]
    return img2

def grayLevelIntensity(img): #0 siyah
    img2 = np.zeros((img.shape[0], img.shape[1]))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if(img[i,j]-70 < 0):
                img2[i,j] = 0
            else:
                img2[i,j] = img[i,j] - 70
    return img2

img1 = plt.imread('Lion.jpg')
grayLevel = RGBtoGrayLevel(img1)
grayLevelIntensity = grayLevelIntensity(grayLevel)
redIntensity = redIntensity(img1)

plt.imshow(img1)
plt.show()

plt.imshow(grayLevel, plt.cm.gray)
plt.show() 

plt.imshow(grayLevelIntensity, plt.cm.gray)
plt.show()

plt.imshow(redIntensity)
plt.show()
```
